```python
import pandas as pd
df = pd.read_csv('/content/Titanic.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
from mlxtend.frequent_patterns import association_rules, apriori
```

# **Pre-processing**

As the data is not in transaction format we will first use get_dummies on columns.

```python
df = pd.get_dummies(df)
```

```python
df.head()
```

# **Apriori Algorithm**

```python
frequent_itemsets = apriori(df, min_support = 0.1, use_colnames=True)
frequent_itemsets
```

This will display us the combinations and cols whose support is > 0.1.

```python
rule = association_rules(frequent_itemsets, metric = "lift", min_threshold = .70)
rule
```

A leverage value of 0 indicated independence. Range will be [-1,1]

A high conviction values means consequent is highly dependent on antecedent.

```python
rule.sort_values('lift', ascending=False)[0:20]
```

```python
rule[rule.lift>1]
```
